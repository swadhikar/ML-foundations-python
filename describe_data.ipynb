```python
import pandas as pd
import os
```

```python
data_path = './data/washers.csv'

if os.path.exists(data_path):
    print('File Exists')
else:
    print('Unable to locate file')
```

```python
washers = pd.read_csv(data_path)
```

```python
washers.head()
```

```python
washers.describe()
# washers.describe()[['WaterUse']]
```

```python
washers[ ['BrandName'] ].value_counts()
```

```python
washers[ ['BrandName'] ].value_counts(normalize=True)
```

```python
washers[['Volume']].mean()
```

```python
# How to get Group-level Aggregations
washers.groupby('BrandName')[['Volume']].mean()
```

```python
washers.groupby('BrandName')[['WaterUse']].mean()
```

```python
washers.groupby('BrandName')[['WaterUse']].mean().sort_values(by='WaterUse')
```

```python
washers.groupby('BrandName')[['Volume']].agg(['min', 'max', 'mean', 'median'])
```
